# Notebook 02: Breadth-First Search (BFS) and Depth-First Search (DFS)

## Overview

This notebook explores two fundamental graph traversal algorithms:
- **Breadth-First Search (BFS)**: Explores nodes level by level, guaranteeing the shortest path in unweighted graphs
- **Depth-First Search (DFS)**: Explores as deep as possible along each branch before backtracking

## Learning Objectives

By the end of this notebook, you will:
1. Understand how BFS and DFS explore the search space differently
2. Recognize when BFS guarantees optimal paths vs. when DFS does not
3. Compare their performance characteristics
4. Visualize their exploration patterns

## Key Concepts

### BFS (Breadth-First Search)
- Uses a **queue** (FIFO: First In, First Out)
- Explores all neighbors at distance *k* before exploring neighbors at distance *k+1*
- **Guarantees shortest path** in terms of number of steps (unweighted graphs)
- Time Complexity: O(V + E) where V = vertices, E = edges
- Space Complexity: O(V) for the queue

### DFS (Depth-First Search)
- Uses a **stack** (LIFO: Last In, First Out)
- Explores as far as possible along each branch before backtracking
- **Does NOT guarantee shortest path**
- Time Complexity: O(V + E)
- Space Complexity: O(V) for the stack (or recursion depth)

## Part 1: Simple Grid Comparison

Let's start with a simple grid to see how BFS and DFS behave differently.

```python
# Create a simple 15x15 grid with light obstacles
grid = Grid(width=15, height=15, obstacle_density=0.15, 
            movement_mode=MovementMode.FOUR_DIRECTIONAL, random_seed=42)

# Define start and goal positions
start = (1, 1)
goal = (13, 13)

# Generate obstacles
grid.generate_obstacles(start, goal)

print(f"Grid created: {grid.width}x{grid.height}")
print(f"Obstacles: {len(grid.obstacles)}")
print(f"Start: {start}, Goal: {goal}")
```

### Run BFS

```python
# Run BFS
bfs_result = bfs(grid, start, goal)

# Display results
print(f"Algorithm: {bfs_result.algorithm_name}")
print(f"Success: {bfs_result.success}")
print(f"Path length: {bfs_result.path_length}")
print(f"Path cost: {bfs_result.path_cost:.2f}")
print(f"Nodes visited: {bfs_result.nodes_visited}")
print(f"Runtime: {bfs_result.runtime_ms:.2f} ms")
print(f"Message: {bfs_result.message}")
```

```python
# Visualize BFS result
fig = create_grid_plot(grid, start, goal, bfs_result, figsize=(12, 10))
plt.show()
```

### Run DFS

Now let's run DFS on the same grid and compare.

```python
# Run DFS
dfs_result = dfs(grid, start, goal)

# Display results
print(f"Algorithm: {dfs_result.algorithm_name}")
print(f"Success: {dfs_result.success}")
print(f"Path length: {dfs_result.path_length}")
print(f"Path cost: {dfs_result.path_cost:.2f}")
print(f"Nodes visited: {dfs_result.nodes_visited}")
print(f"Runtime: {dfs_result.runtime_ms:.2f} ms")
print(f"Message: {dfs_result.message}")
```

```python
# Visualize DFS result
fig = create_grid_plot(grid, start, goal, dfs_result, figsize=(12, 10))
plt.show()
```

### Side-by-Side Comparison

```python
# Create side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Helper function to plot on specific axis
def plot_result_on_axis(ax, grid, start, goal, result, title):
    from matplotlib.colors import ListedColormap
    from matplotlib.patches import Patch
    
    # Create visualization grid
    vis_grid = np.zeros((grid.height, grid.width))
    
    # Mark obstacles
    for obstacle in grid.obstacles:
        row, col = obstacle
        vis_grid[row, col] = 1
    
    # Mark visited nodes
    if result and result.visited_order:
        for pos in result.visited_order:
            row, col = pos
            if pos != start and pos != goal:
                vis_grid[row, col] = 2
    
    # Mark path
    if result and result.path:
        for pos in result.path:
            row, col = pos
            if pos != start and pos != goal:
                vis_grid[row, col] = 3
    
    # Mark start and goal
    vis_grid[start[0], start[1]] = 4
    vis_grid[goal[0], goal[1]] = 5
    
    # Plot
    colors = ['white', 'black', 'lightblue', 'yellow', 'green', 'red']
    cmap = ListedColormap(colors)
    ax.imshow(vis_grid, cmap=cmap, vmin=0, vmax=5)
    
    # Grid lines
    ax.set_xticks(np.arange(-0.5, grid.width, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, grid.height, 1), minor=True)
    ax.grid(which='minor', color='gray', linestyle='-', linewidth=0.5)
    ax.tick_params(which='minor', size=0)
    
    # Title with stats
    stats = f"Path: {result.path_length}, Visited: {result.nodes_visited}, Time: {result.runtime_ms:.2f}ms"
    ax.set_title(f"{title}\n{stats}", fontsize=12, fontweight='bold')

# Plot both results
plot_result_on_axis(axes[0], grid, start, goal, bfs_result, "BFS (Breadth-First Search)")
plot_result_on_axis(axes[1], grid, start, goal, dfs_result, "DFS (Depth-First Search)")

plt.tight_layout()
plt.show()

# Print comparison
print("\n" + "="*60)
print("COMPARISON SUMMARY")
print("="*60)
print(f"{'Metric':<25} {'BFS':<15} {'DFS':<15}")
print("-"*60)
print(f"{'Path Length':<25} {bfs_result.path_length:<15} {dfs_result.path_length:<15}")
print(f"{'Nodes Visited':<25} {bfs_result.nodes_visited:<15} {dfs_result.nodes_visited:<15}")
print(f"{'Runtime (ms)':<25} {bfs_result.runtime_ms:<15.2f} {dfs_result.runtime_ms:<15.2f}")
print(f"{'Optimal Path':<25} {'Yes':<15} {'No':<15}")
print("="*60)
```

## Part 2: Maze with Narrow Corridors

Let's test both algorithms on a more challenging scenario: a maze with narrow corridors. This demonstrates when DFS can sometimes be faster but finds suboptimal paths.

```python
# Create a grid with corridor-like obstacles
maze_grid = Grid(width=20, height=20, obstacle_density=0.0, 
                 movement_mode=MovementMode.FOUR_DIRECTIONAL)

# Manually create a maze pattern
# Vertical walls
for row in range(2, 18):
    maze_grid.add_obstacle((row, 5))
    maze_grid.add_obstacle((row, 10))
    maze_grid.add_obstacle((row, 15))

# Horizontal walls with gaps
for col in range(1, 5):
    maze_grid.add_obstacle((8, col))
for col in range(6, 10):
    maze_grid.add_obstacle((12, col))
for col in range(11, 15):
    maze_grid.add_obstacle((8, col))
for col in range(16, 19):
    maze_grid.add_obstacle((12, col))

maze_start = (1, 1)
maze_goal = (18, 18)

print(f"Maze grid created: {maze_grid.width}x{maze_grid.height}")
print(f"Obstacles: {len(maze_grid.obstacles)}")
```

```python
# Run both algorithms on the maze
maze_bfs_result = bfs(maze_grid, maze_start, maze_goal)
maze_dfs_result = dfs(maze_grid, maze_start, maze_goal)

# Visualize side by side
fig, axes = plt.subplots(1, 2, figsize=(20, 8))
plot_result_on_axis(axes[0], maze_grid, maze_start, maze_goal, maze_bfs_result, "BFS on Maze")
plot_result_on_axis(axes[1], maze_grid, maze_start, maze_goal, maze_dfs_result, "DFS on Maze")
plt.tight_layout()
plt.show()

print(f"\nBFS Path Length: {maze_bfs_result.path_length}")
print(f"DFS Path Length: {maze_dfs_result.path_length}")
print(f"\nPath difference: {abs(maze_dfs_result.path_length - maze_bfs_result.path_length)} steps")
```

## Part 3: Performance Analysis

Let's benchmark both algorithms across different grid sizes to understand their performance characteristics.

```python
# Test different grid sizes
grid_sizes = [10, 15, 20, 25, 30]
bfs_times = []
dfs_times = []
bfs_visited = []
dfs_visited = []

for size in grid_sizes:
    # Create grid
    test_grid = Grid(width=size, height=size, obstacle_density=0.2, 
                     movement_mode=MovementMode.FOUR_DIRECTIONAL, random_seed=42)
    test_start = (1, 1)
    test_goal = (size-2, size-2)
    test_grid.generate_obstacles(test_start, test_goal)
    
    # Run BFS
    result_bfs = bfs(test_grid, test_start, test_goal)
    bfs_times.append(result_bfs.runtime_ms)
    bfs_visited.append(result_bfs.nodes_visited)
    
    # Run DFS
    result_dfs = dfs(test_grid, test_start, test_goal)
    dfs_times.append(result_dfs.runtime_ms)
    dfs_visited.append(result_dfs.nodes_visited)
    
    print(f"Grid {size}x{size}: BFS={result_bfs.runtime_ms:.2f}ms, DFS={result_dfs.runtime_ms:.2f}ms")

print("\nBenchmarking complete!")
```

```python
# Plot performance comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Runtime comparison
axes[0].plot(grid_sizes, bfs_times, marker='o', linewidth=2, markersize=8, label='BFS')
axes[0].plot(grid_sizes, dfs_times, marker='s', linewidth=2, markersize=8, label='DFS')
axes[0].set_xlabel('Grid Size (N x N)', fontsize=12)
axes[0].set_ylabel('Runtime (ms)', fontsize=12)
axes[0].set_title('Runtime Comparison', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# Nodes visited comparison
axes[1].plot(grid_sizes, bfs_visited, marker='o', linewidth=2, markersize=8, label='BFS')
axes[1].plot(grid_sizes, dfs_visited, marker='s', linewidth=2, markersize=8, label='DFS')
axes[1].set_xlabel('Grid Size (N x N)', fontsize=12)
axes[1].set_ylabel('Nodes Visited', fontsize=12)
axes[1].set_title('Nodes Visited Comparison', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Key Takeaways

### When to Use BFS:
- When you need the **shortest path** (in terms of steps)
- For unweighted graphs
- When the goal is relatively close to the start
- In level-based games or puzzles

### When to Use DFS:
- When any path will do (not necessarily shortest)
- For maze generation
- For topological sorting
- When memory is very limited (recursive DFS uses call stack)
- **Not recommended** for shortest path problems

### Key Differences:
1. **Completeness**: Both are complete (find a solution if one exists)
2. **Optimality**: BFS is optimal for unweighted graphs, DFS is not
3. **Memory**: BFS typically uses more memory (stores entire frontier)
4. **Exploration Pattern**: BFS expands in circles, DFS explores deep branches
