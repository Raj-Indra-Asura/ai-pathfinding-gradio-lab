## Setup and Imports

```python
# Add parent directory to path for imports
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from time import perf_counter

import pandas as pd

from pathfinding_lab.algorithms.astar import astar
from pathfinding_lab.algorithms.bfs import bfs
from pathfinding_lab.core.grid import Grid
from pathfinding_lab.core.types import MovementMode, Position
from pathfinding_lab.heuristics.manhattan import manhattan_distance
```

## Part 1: Build a Typed Helper for Repeated Experiments

```python
def benchmark_once(grid: Grid, start: Position, goal: Position) -> dict[str, float]:
    t0 = perf_counter()
    bfs_result = bfs(grid, start, goal)
    t1 = perf_counter()

    astar_result = astar(grid, start, goal, manhattan_distance)
    t2 = perf_counter()

    return {
        "bfs_runtime_ms": (t1 - t0) * 1000,
        "astar_runtime_ms": (t2 - t1) * 1000,
        "bfs_nodes": float(bfs_result.nodes_visited),
        "astar_nodes": float(astar_result.nodes_visited),
    }
```

## Part 2: Run a Small Performance Sample

```python
rows = []
for seed in [10, 11, 12, 13, 14]:
    grid = Grid(
        width=25,
        height=25,
        obstacle_density=0.2,
        movement_mode=MovementMode.FOUR_DIRECTIONAL,
        random_seed=seed,
    )
    start, goal = (0, 0), (24, 24)
    grid.generate_obstacles(start, goal)
    rows.append(benchmark_once(grid, start, goal))

perf_df = pd.DataFrame(rows)
perf_df
```

```python
perf_df.mean(numeric_only=True).to_frame(name="mean")
```
