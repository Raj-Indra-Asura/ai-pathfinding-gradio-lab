## Part 1: Creating a Basic Grid

Let's start by creating a simple grid. A grid is defined by:
- **Width**: Number of columns
- **Height**: Number of rows
- **Obstacle Density**: Percentage of cells that are obstacles (0.0 to 1.0)
- **Movement Mode**: How entities can move (4-directional or 8-directional)

```python
# Create a 10x10 grid with no obstacles
grid = Grid(width=10, height=10, obstacle_density=0.0, 
            movement_mode=MovementMode.FOUR_DIRECTIONAL)

# Verify dimensions
print(f"Grid size: {grid.width}x{grid.height}")
print(f"Number of obstacles: {len(grid.obstacles)}")
print(f"Movement mode: {grid.movement_mode}")
print(f"Total cells: {grid.width * grid.height}")
```

### Understanding Coordinates

In our grid system:
- Positions are represented as `(row, col)` tuples
- `row` increases going **down** (0 is top)
- `col` increases going **right** (0 is left)
- Origin `(0, 0)` is the **top-left** corner

```python
# Example positions
top_left = (0, 0)
top_right = (0, grid.width - 1)
bottom_left = (grid.height - 1, 0)
bottom_right = (grid.height - 1, grid.width - 1)
center = (grid.height // 2, grid.width // 2)

print(f"Top-left corner: {top_left}")
print(f"Top-right corner: {top_right}")
print(f"Bottom-left corner: {bottom_left}")
print(f"Bottom-right corner: {bottom_right}")
print(f"Center: {center}")
```

## Part 2: Adding Obstacles

Obstacles are cells that cannot be traversed. Let's add some obstacles manually.

```python
# Create a fresh grid
grid = Grid(width=10, height=10, obstacle_density=0.0, 
            movement_mode=MovementMode.FOUR_DIRECTIONAL)

# Add obstacles manually to create a vertical wall
for row in range(2, 8):
    grid.add_obstacle((row, 5))

print(f"Total obstacles: {len(grid.obstacles)}")
print(f"Obstacle positions: {sorted(grid.obstacles)}")
```

### Visualizing the Grid

Let's visualize our grid with obstacles, a start position, and a goal position.

```python
# Define start and goal positions
start = (1, 1)
goal = (8, 8)

# Create visualization
fig = create_grid_plot(grid, start, goal, figsize=(10, 10))
plt.show()

print(f"\nStart position: {start}")
print(f"Goal position: {goal}")
```

## Part 4: Neighbor Generation

A crucial concept in pathfinding is understanding how neighbors are generated. This determines which cells an agent can move to from a given position.

### 4-Directional Movement

In 4-directional movement, you can move:
- Up (decrease row)
- Down (increase row)
- Left (decrease col)
- Right (increase col)

```python
# Create a grid with 4-directional movement
grid_4dir = Grid(width=10, height=10, obstacle_density=0.0, 
                 movement_mode=MovementMode.FOUR_DIRECTIONAL)

# Get neighbors of the center position
position = (5, 5)
neighbors = grid_4dir.get_neighbors(position)

print(f"Position: {position}")
print(f"Movement mode: {grid_4dir.movement_mode}")
print(f"Number of neighbors: {len(neighbors)}")
print(f"\nNeighbors:")
for neighbor in neighbors:
    row_diff = neighbor[0] - position[0]
    col_diff = neighbor[1] - position[1]
    direction = ""
    if row_diff == -1: direction = "Up"
    elif row_diff == 1: direction = "Down"
    elif col_diff == -1: direction = "Left"
    elif col_diff == 1: direction = "Right"
    print(f"  {neighbor} - {direction}")
```

### 8-Directional Movement

In 8-directional movement, you can also move diagonally:
- Up-Left, Up-Right
- Down-Left, Down-Right

```python
# Create a grid with 8-directional movement
grid_8dir = Grid(width=10, height=10, obstacle_density=0.0, 
                 movement_mode=MovementMode.EIGHT_DIRECTIONAL)

# Get neighbors of the center position
position = (5, 5)
neighbors = grid_8dir.get_neighbors(position)

print(f"Position: {position}")
print(f"Movement mode: {grid_8dir.movement_mode}")
print(f"Number of neighbors: {len(neighbors)}")
print(f"\nNeighbors:")
for neighbor in neighbors:
    row_diff = neighbor[0] - position[0]
    col_diff = neighbor[1] - position[1]
    direction = ""
    if row_diff == -1 and col_diff == -1: direction = "Up-Left"
    elif row_diff == -1 and col_diff == 0: direction = "Up"
    elif row_diff == -1 and col_diff == 1: direction = "Up-Right"
    elif row_diff == 0 and col_diff == -1: direction = "Left"
    elif row_diff == 0 and col_diff == 1: direction = "Right"
    elif row_diff == 1 and col_diff == -1: direction = "Down-Left"
    elif row_diff == 1 and col_diff == 0: direction = "Down"
    elif row_diff == 1 and col_diff == 1: direction = "Down-Right"
    print(f"  {neighbor} - {direction}")
```

### Edge and Corner Cases

What happens at the edges and corners of the grid?

```python
# Test corner position
corner = (0, 0)
corner_neighbors_4dir = grid_4dir.get_neighbors(corner)
corner_neighbors_8dir = grid_8dir.get_neighbors(corner)

print(f"Corner position: {corner}")
print(f"4-directional neighbors: {len(corner_neighbors_4dir)} - {sorted(corner_neighbors_4dir)}")
print(f"8-directional neighbors: {len(corner_neighbors_8dir)} - {sorted(corner_neighbors_8dir)}")

# Test edge position
edge = (0, 5)
edge_neighbors_4dir = grid_4dir.get_neighbors(edge)
edge_neighbors_8dir = grid_8dir.get_neighbors(edge)

print(f"\nEdge position: {edge}")
print(f"4-directional neighbors: {len(edge_neighbors_4dir)} - {sorted(edge_neighbors_4dir)}")
print(f"8-directional neighbors: {len(edge_neighbors_8dir)} - {sorted(edge_neighbors_8dir)}")
```

### Neighbors with Obstacles

Obstacles block neighbors. Let's see this in action.

```python
# Create a grid and add obstacles around a position
grid_blocked = Grid(width=10, height=10, obstacle_density=0.0, 
                    movement_mode=MovementMode.FOUR_DIRECTIONAL)

center = (5, 5)

# Add obstacles on two sides
grid_blocked.add_obstacle((4, 5))  # Above
grid_blocked.add_obstacle((5, 6))  # Right

# Get neighbors
available_neighbors = grid_blocked.get_neighbors(center)

print(f"Position: {center}")
print(f"Obstacles: {sorted(grid_blocked.obstacles)}")
print(f"Available neighbors: {len(available_neighbors)} - {sorted(available_neighbors)}")
print(f"\nBlocked directions: Up, Right")
print(f"Available directions: Down, Left")
```

## Part 5: Checking Cell Properties

The Grid class provides methods to check if a position is valid, is an obstacle, etc.

```python
# Create a grid with some obstacles
grid = Grid(width=10, height=10, obstacle_density=0.0, 
            movement_mode=MovementMode.FOUR_DIRECTIONAL)
grid.add_obstacle((5, 5))

# Test various positions
test_positions = [
    (5, 5),    # Obstacle
    (0, 0),    # Valid, top-left
    (-1, 0),   # Out of bounds (negative)
    (10, 10),  # Out of bounds (too large)
    (7, 7),    # Valid, empty cell
]

print("Position\t\tIn Bounds\tIs Obstacle")
print("-" * 50)
for pos in test_positions:
    in_bounds = 0 <= pos[0] < grid.height and 0 <= pos[1] < grid.width
    is_obstacle = pos in grid.obstacles
    print(f"{str(pos):<20}\t{in_bounds}\t\t{is_obstacle}")
```

## Key Takeaways

### Grid Fundamentals:
1. **Coordinates**: Grids use `(row, col)` indexing with `(0, 0)` at top-left
2. **Obstacles**: Cells that cannot be traversed
3. **Movement Modes**: Determine how many neighbors each cell has
   - 4-directional: Up, Down, Left, Right (max 4 neighbors)
   - 8-directional: Includes diagonals (max 8 neighbors)

### Important Concepts:
- Edge and corner cells have fewer neighbors
- Obstacles reduce the number of available neighbors
- Random obstacle generation ensures start/goal accessibility
- Visualization helps understand grid structure

### Grid Class Methods:
- `add_obstacle(pos)`: Add an obstacle at position
- `remove_obstacle(pos)`: Remove an obstacle
- `generate_obstacles(start, goal)`: Create random obstacles
- `get_neighbors(pos)`: Get valid neighboring positions
