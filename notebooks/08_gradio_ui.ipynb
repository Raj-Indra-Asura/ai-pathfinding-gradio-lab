## Part 2: Build a Small Gradio Blocks Demo

```python
def summarize_settings(algorithm: str, heuristic: str, width: int, height: int) -> str:
    return (
        f"Algorithm: {algorithm} | Heuristic: {heuristic} | "
        f"Grid: {width}x{height}"
    )

with gr.Blocks() as mini_demo:
    gr.Markdown("### Mini Config Summary Demo")
    algorithm = gr.Dropdown(["BFS", "Dijkstra", "A*"], value="A*", label="Algorithm")
    heuristic = gr.Dropdown(["Manhattan", "Euclidean", "Octile"], value="Manhattan", label="Heuristic")
    width = gr.Slider(10, 40, value=20, step=1, label="Width")
    height = gr.Slider(10, 40, value=20, step=1, label="Height")
    output = gr.Textbox(label="Summary")
    gr.Button("Summarize").click(
        summarize_settings,
        inputs=[algorithm, heuristic, width, height],
        outputs=[output],
    )

mini_demo
```
